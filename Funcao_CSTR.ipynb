```python
from dataclasses import dataclass

@dataclass
class ParametroReator:
  """
  F: Vazão de alimentação em L/min
  V: Volume do reator em L
  k: constante de velocidade em min^-1
  """
  F: float
  V: float
  k: float
```

```python
def modelo_CSTR(Cain, Parametro):
  F = Parametro.F
  V = Parametro.V
  k = Parametro.k
  Ca = F/(F+V*k)*Cain
  return Ca
```

```python
import numpy as np

def simula_cstr(n):
  Cao = 1.0
  Parametro = ParametroReator(F=50, V=100, k=0.5)
  Calist = [Cao]
  Xlist = []
  for i in range(n):
    Ca = modelo_CSTR(Calist[i], Parametro)
    X = ((Calist[0]-Ca)/Calist[0])*100
    Calist.append(Ca)
    Xlist.append(X)
  Calist.pop(0)
  return Calist, Xlist
```
