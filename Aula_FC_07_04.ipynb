```python
nivel_atual = 50
nivel_min = 10
nivel_max = 90

try:
  if not isinstance(nivel_atual, (int,float)):
    raise TypeError('Tipo de dado inválido')

  if nivel_atual < nivel_min:
   mensagem = 'Atenção, nivel muito baixo'

  elif nivel_atual > nivel_max:
    mensagem = 'Atenção, nivel muito alto'

  else:
    mensagem = 'Nivel normal'

except ValueError as e:
  mensagem = (f"Tipo de dado inválido: {e}")

print(mensagem)
```
