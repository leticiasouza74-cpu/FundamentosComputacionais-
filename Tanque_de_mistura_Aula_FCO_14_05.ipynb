```python
from dataclasses import dataclass

@dataclass
class Corrente:
  """
  Vazão: Vazão da Corrente em kg/h
  Fração: Fração mássica da Corrente em kg/kg
  """
  vazao: float
  fracao: float

```

```python
def misturador(c1: Corrente, c2: Corrente):

  '''
  Calcule a vazão e a fração mássica da mistura de  correntes.
  Entradas:
    c1: Corrente 1
    c2: Corrente 2
  Saída:
    F: Vazão da mistura em kg/h
    x: Fração mássica da mistura em kg/kg
  '''
  Fa = c1.vazao
  xa = c1.fracao

  Fb = c2.vazao
  xb = c2.fracao

  F = Fa + Fb
  x = (Fa*xa + Fb*xb)/F

  return F, x

```
