```python
import numpy as np

#print(conc)

# Inicializa os contadores para cada classe
baixa, adequada, alta = 0, 0,0
i = 0

# Loop para realizar 10 leituras consecutivas
for i in range(10):

  # Gera um valor aleatório de concentração entre 0 e 10
  conc = np.random.rand()*10

  # Verifica em qual faixa a concentração se encontra
  if conc<3:
    classe = "Baixa"
    baixa += 1

  elif conc<=7:
    classe = 'Adequada'
    adequada += 1
  else:
    classe = 'Alta'
    alta += 1

  print(f"Leitura {i+1}: Concentração = {conc:.1f} mol/L -> {classe}")

# Exibe o número de ocorrências de cada classe
print("\n Ocorrências:")
print(f"  Baixa: {baixa}")
print(f"  Adequada: {adequada}")
print(f"  Alta: {alta}")

# Cria uma lista com as ocorrências
ocorrencias = [baixa, adequada, alta]
major = max(ocorrencias)

#Verifica a classe do major
if major == baixa:
    print("\nCondição predominante: Baixa")
elif major == adequada:
    print("\nCondição predominante: Adequada")
else:
    print("\nCondição predominante: Alta")
```
